```python
cons = pd.read_csv('https://als-hiring.s3.amazonaws.com/fake_data/2020-07-01_17%3A11%3A00/cons.csv')
email = pd.read_csv('https://als-hiring.s3.amazonaws.com/fake_data/2020-07-01_17%3A11%3A00/cons_email.csv')
chapter_sub = pd.read_csv('https://als-hiring.s3.amazonaws.com/fake_data/2020-07-01_17%3A11%3A00/cons_email_chapter_subscription.csv')
```

```python
#desired table <br>
#Column	    Type	    Description + table location
#email	    string	    Primary email address - email in email table
#code	    string	    Source code - source column in con table
#is_unsub	boolean	    Is the primary email address unsubscribed? - isunsub in chapter table
#created_dt	datetime	Person creation datetime - create_dt in cons
#updated_dt	datetime	Person updated datetime - modified_dt in cons

#how to link tables
#cons to email = cons_id to cons_id
#email to chapter_sub = cons_email_id to cons_email_id
```

```python
# filter down to only chapter IDs that are 1
chapter_sub = chapter_sub[chapter_sub['chapter_id']==1]
```

```python
# subset tables down to just the rows that are needed
e2 = email[['cons_id','cons_email_id','email']]
c2 = cons[['cons_id','source','create_dt','modified_dt']]
cs2 = chapter_sub[['cons_email_id','isunsub']]
```

```python
print(e2.shape, c2.shape, cs2.shape)
```

```python
# merge tables together
# use inner joins so that tables only include full information
ec2 = pd.merge(e2, c2, on='cons_id', how='inner')
tab = pd.merge(ec2, cs2, on='cons_email_id', how='inner')
```

```python
# drop cons_id, cons_email
tab = tab[['email','source','isunsub','create_dt','modified_dt']]
print(tab.head(), tab.info())
```

```python
# change column names
tab = tab.rename(columns={"source": "code", "isunsub": "is_unsub", "create_dt":"created_dt","modified_dt":"updated_dt"})
```

```python
tab.info()
```

```python
#tab.astype({'cons_id': 'int32'}).dtypes
# change data types
tab = tab.astype({'email':str,'code':str, 'is_unsub':bool})
tab[['created_dt','updated_dt']] = tab[['created_dt','updated_dt']].apply(pd.to_datetime)
```

```python
# drop publucates just in case
tab = tab.drop_duplicates()
print(tab.head(),tab.info())
```

```python
# working directory set at beginning, table will write to working directory
tab.to_csv('people.csv', index=False, header=working_directory)
```

```python
# exercise 2
#tab.groupby('created_dt')['email'].sum()
acq_series = tab.groupby([tab['created_dt'].dt.date])['email'].count()
# make the series into a data frame
acq_table = pd.DataFrame({'acquisition_dt':acq_series.index, 'acquisitions':acq_series.values})
```

```python
# rename columns
acq_table['acquisition_dt'] = acq_table[['acquisition_dt']].apply(pd.to_datetime)
print(acq_table.info(), acq_table.head())
```

```python
# save - will write to working directory
acq_table.to_csv('acquisition.csv', index=False)
```
